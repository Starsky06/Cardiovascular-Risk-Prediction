## 4-Data Pre-processing

#### Separating input matrix from output vector

```python
x_train = data.drop('Cardiovascular_risk(y)', axis = 1)
y_train = data[['Cardiovascular_risk(y)']]
print(x_train)
print(y_train)

 
```

#### Handling missing values

```python
x_train = x_train.fillna(0)
y_train = y_train.fillna(0)
print('x_train.shape: ' , x_train.shape)
print('y_train.shape: ' , y_train.shape)
```

#### Split the input matrix into numerical and categorical sets

```python
x_train_num = x_train[['Age', 'Height(cm)', 'Weight(kg)' ,'Vege_day','Meals_day','Water_intake(L)','Exercise','Income']]
x_train_cat = x_train[['Gender', 'Family_history','Alcohol', 'Junk_food','Snack', 'Smoking','Transportation', 'TV' ,'Discipline']]
print(x_train_num)
print(x_train_cat)
```

```python
print('x_train_num shape: ' , x_train_num.shape)
print('x_train_cat shape: ' , x_train_cat.shape)
print('y_train shape: ' , y_train.shape)
```

### Preprocessing Numerical Data

#### Standardization

```python
from sklearn.preprocessing import StandardScaler
#StandardScaler
scaler = StandardScaler()
x_train_num_tr = scaler.fit_transform(x_train_num)

print('Type of X_num:',         type(x_train_num))
print('Type of X_num_tr',       type(x_train_num_tr))
print('Mean of all columns:',   x_train_num_tr.mean())
print('Std of all columns:',    x_train_num_tr.std())
```

### Preprocessing Categorical Data

###### Nominal data, Ordinal data

```python
from sklearn.preprocessing import OrdinalEncoder

x_train_cat_ordinal = x_train_cat[['Alcohol', 'Snack', 'TV', 'Transportation']]
x_train_cat_nominal = x_train_cat.drop(['Alcohol', 'Snack', 'TV','Transportation' ], axis = 1)
selected_columns = ['Gender', 'Family_history', 'Junk_food','Smoking', 'Discipline']

x_train_cat_nominal_tr = pd.get_dummies(x_train_cat_nominal[selected_columns], drop_first=True)

#OrdinalEncoder
encoder = OrdinalEncoder()
x_train_cat_ordinal_tr = encoder.fit_transform(x_train_cat_ordinal)

print(x_train_cat_nominal_tr.shape)
print(x_train_cat_ordinal_tr.shape)
```

#### Finalizing training set

```python
#Finalizing training set
x_train_tr = np.hstack([x_train_cat_nominal_tr, x_train_cat_ordinal_tr, x_train_num_tr])
x_train_tr
```

###  Preprocessing output vector (y)

```python
from sklearn.preprocessing import LabelEncoder

#LabelEncoder
label_encoder = LabelEncoder()

cardiovascular_risk = data['Cardiovascular_risk(y)'].values.ravel()

y_train_tr = label_encoder.fit_transform(cardiovascular_risk)


y_train_tr = np.array(y_train_tr)

y_train_tr
```

```python
#print shape of finalizing training set

print('x_train_tr shape: ' , x_train_tr.shape)
print('y_train_tr shape: ' , y_train_tr.shape)
```

### Split X and y into training and testing sets

```python
from sklearn.model_selection import train_test_split

#X_train, x_test, Y_train, y_test which is name of split test and train data
X_train, x_test, Y_train, y_test = train_test_split(x_train_tr, y_train_tr, test_size = 0.2, random_state = 30)

print("Full X Shape:", x_train_tr.shape)
print("X Train Shape:",X_train.shape)
print("X Test Shape:",x_test.shape)
print("Y Train Shape:",Y_train.shape)
print("Y Test Shape:",y_test.shape)
```

## 5-Model Training and Validation

### Model Training

### 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#LogisticRegression
log_reg = LogisticRegression(solver='liblinear')
log_reg.fit(X_train, Y_train)

```

### Model Validation

#### Define a function of show result

```python
def show10results(y_train, y_pred, num=10):
    print("Result for 10 random samples: ")
    for i in range(num):
        selected = np.random.randint(0, len(y_train))
        print("Actual = {} Predict = {}".format(y_train[selected], y_pred[selected]))
```

```python
show10results(Y_train, y_pred, num=10)
```

#### Training Accuracy ,Confusion Matrix, Classification Report (Data vs Prediction)

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

train_accuracy = accuracy_score(Y_train, y_pred)
confusion_matrix = confusion_matrix(Y_train, y_pred)
classification_report = classification_report(Y_train, y_pred)

print("Training Accuracy: ", train_accuracy)  #accurancy of logistic Regression

print("Confusion Matrix: \n", confusion_matrix)

print("Classification Report (Training):\n", classification_report)

print("Unique classes in Y_train:", np.unique(Y_train))
```

#### Cross_val_predict (divide into 5 parts to predict)

```python
from sklearn.model_selection import cross_val_predict
y_pred_cv = cross_val_predict(log_reg, X_train, Y_train, cv = 5)

show10results(Y_train, y_pred_cv)

print("y_pred_cv.shape :",y_pred_cv.shape)
print("Y_train.shape :", Y_train.shape)
```

#### Precision Score , Recall Score , F1 Score

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

precision_scr = precision_score(Y_train, y_pred_cv, average = 'weighted' ) # default is binary as the dataset is more than two classfication
recall_scr = recall_score(Y_train, y_pred_cv, average = 'weighted' )
f1_scr = f1_score(Y_train, y_pred_cv, average = 'weighted') #none, micro, macro, weighted 

print('Precision Score: ', precision_scr)
print('Recall Score: ', recall_scr)
print('F1 Score: ', f1_scr)

print("y_pred_cv.shape :",y_pred_cv.shape)
print("Y_train.shape :", Y_train.shape)
```

#### Curve for precision-recall 

```python

from sklearn.metrics import precision_recall_curve
from sklearn.preprocessing import label_binarize

#Binarize the labels for multiclass
Y_train_bin = label_binarize(Y_train, classes=[0, 1, 2]) # three category
n_classes = Y_train_bin.shape[1]

#Calculate the precision and recall for each class
for i in range(n_classes):
    precisions, recalls, _ = precision_recall_curve(Y_train_bin[:, i], y_scores_cv[:, i])
    plt.plot(recalls, precisions, label=f'Class {i}')

#Draw a diagonal line
plt.plot(np.linspace(0, 1, 20), np.linspace(1, 0, 20), 'k--')

#Marking a graphic
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Training Set)")
plt.legend(loc="best")
plt.axis([0, 1, 0, 1])
plt.show()
```

#### Curve for ROC

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

#Binarize the labels for multiclass
Y_train_bin = label_binarize(Y_train, classes=[0, 1, 2])  
n_classes = Y_train_bin.shape[1]

#Calculate the ROC curve for each class
plt.figure(figsize=(8, 6))

for i in range(n_classes):
    #Calculating ROC Curve
    fpr, tpr, _ = roc_curve(Y_train_bin[:, i], y_scores_cv[:, i])
    #Calculate AUC (area under the curve)
    roc_auc = auc(fpr, tpr)
    #Plotting the ROC curve
    plt.plot(fpr, tpr, lw=2, label=f'Class {i} (AUC = {roc_auc:.2f})')

#Draw a diagonal line (a random guess at the baseline)
plt.plot([0, 1], [0, 1], 'k--')

#Set the title and label of the graph
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='best')
plt.axis([0, 1, 0, 1])
plt.show()
```

## 6- Model Tuning and Testing

### Model Tuning

#### RandomizedSearchCV

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform

# Define parameter grid with solver-specific restrictions
param_grid = [
    {'solver': ['newton-cg'], 'penalty': ['none', 'l2'], 'C': loguniform(1e-5, 100)},
    {'solver': ['lbfgs'], 'penalty': ['none', 'l2'], 'C': loguniform(1e-5, 100)},
    {'solver': ['liblinear'], 'penalty': ['l1', 'l2'], 'C': loguniform(1e-5, 100)}
]

# Define the model
log_reg = LogisticRegression()

# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(log_reg, param_grid, n_iter=500, scoring='accuracy', n_jobs=-1, cv=3, random_state=1)

# Perform random search
print('Performing random search...', end=' ')
random_search.fit(X_train, Y_train)
print('Done performing random search!')

# Get the best model
best_model = random_search.best_estimator_
print(best_model)
```
